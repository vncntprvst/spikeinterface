```python
import numpy as np
import matplotlib.pyplot as plt
import spikeinterface.extractors as se
import spikeinterface.widgets as sw
```

```python
import os
binary_file_path="D:\\Vincent\\vIRt57\\vIRt57_0216\\SpikeSorting\\vIRt57_0216_5732\\"
os.chdir(binary_file_path)
```

```python
filename = "vIRt57_0216_5732_export.bin"
print(f"Reading {filename}")
recording = se.BinaryRecordingExtractor(filename, 
		sampling_frequency=30000.0,
		num_chan = 32,
		dtype='int16', #np.int16, 
		t_starts=None, 
		channel_ids=None, time_axis=0, 
		file_offset=0, gain_to_uV=None, 
		offset_to_uV=None, 
		is_filtered=None)
```

```python
print(f"Plot first two second of recording")
w_ts = sw.plot_timeseries(recording, time_range=(0, 2))
```

```python
# Get data snippet (first two seconds)
samp_freq = recording.get_sampling_frequency()
num_chan = recording.get_num_channels()
trace_snippet = recording.get_traces(start_frame=int(samp_freq * 0), end_frame=int(samp_freq * 2))
print(f"trace_snippet shape is {trace_snippet.shape}")
```

```python
print(f"Plot snippet - channel 0")
plt.plot(trace_snippet[:, 0], label="snippet")
plt.legend()
plt.show()
```
